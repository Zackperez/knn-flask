```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

le = LabelEncoder()

data = pd.read_csv('breast-cancer.csv')
```

```python
data.replace({'node-caps': {'?': 'yes'}}, inplace= True)
data.replace({'breast-quead': {'?': 'central'}}, inplace= True)
```

```python
# dividir el conjunto de datos en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data_new.drop('class', axis=1), data_new['class'], test_size=0.2)
```

```python
# instanciar el clasificador KNN con k=5
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# entrenar el clasificador KNN
knn.fit(X_train, y_train)
```

```python
# predecir el conjunto de prueba
y_pred = knn.predict(X_test)
```

```python
# evaluardatasakjÃ±ldalskasdfasdfsdfsdfsdfssdfsdf
accuracy = knn.score(X_test, y_test)
print("PrecisiÃ³n:", accuracy)
```
