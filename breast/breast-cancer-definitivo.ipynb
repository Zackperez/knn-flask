```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

le = LabelEncoder()

data = pd.read_csv('breast-cancer.csv')
```

```python
data.info()
```

# Limpieza de datos vacios

```python
data.replace({'node-caps': {'?': 'yes'}}, inplace= True)
data.replace({'breast-quead': {'?': 'central'}}, inplace= True)
```

# Copia del dataframe

```python
data_copia = data.copy()
```

```python
data_copia.info()
```

### Reemplazo de la columna 'class'

```python
data_copia['class_numerica'] = le.fit_transform(data_copia['class'])
data_copia['class_menonumero'] = le.fit_transform(data_copia['menopause'])
data_copia['class_nodenumerica'] = le.fit_transform(data_copia['node-caps'])
data_copia['class_breastnumerico'] = le.fit_transform(data_copia['breast'])
data_copia['class_breastnumericoqd'] = le.fit_transform(data_copia['breast-quead'])
```

## EliminaciÃ³n de las columnas originales

```python
data_copia = data_copia.drop('class', axis = 1)
data_copia = data_copia.drop('menopause', axis = 1)
data_copia = data_copia.drop('node-caps', axis = 1)
data_copia = data_copia.drop('breast', axis = 1)
data_copia = data_copia.drop('breast-quead', axis = 1)
```

```python
data_copia.info()
```

# Renombramiento de las columnas

```python
#Columna Class
data_copia.rename(columns={'class_numerica': 'class'}, inplace = True)
#Columna menopause
data_copia.rename(columns={'class_menonumero': 'menopause'}, inplace = True)
#Columna node-caps
data_copia.rename(columns={'class_nodenumerica': 'node-caps'}, inplace = True)
#Columna breast
data_copia.rename(columns={'class_breastnumerico': 'breast'}, inplace = True)
#Columna breast-quead
data_copia.rename(columns={'class_breastnumericoqd': 'breast-quead'}, inplace = True)
```
